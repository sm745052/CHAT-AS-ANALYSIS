```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#compute predicted length distribution and gt length distribution
predlen_dist = {}
gtlen_dist = {}
with open('../eval.mpc', 'r') as f:
    for query in f:
        query = query.strip()
        query = query.split('\t')
        if(query[-1]=='-'):
            continue
        y = query[1]
        pred = query[2]
        predlen_dist[len(pred)] = predlen_dist.get(len(pred), 0) + 1
        gtlen_dist[len(y)] = gtlen_dist.get(len(y), 0) + 1
```

```python
# plt both distributions
x = np.arange(0, 100)
y = []
for i in x:
    y.append(predlen_dist.get(i, 0))
plt.bar(x, y, color='green')
plt.title("Predicted length distribution [MPC]")
plt.xlabel("Predicted length")
plt.ylabel("Count")
plt.show()
y = []
for i in x:
    y.append(gtlen_dist.get(i, 0))
plt.bar(x, y, color='green')
plt.title("Ground truth length distribution [MPC]")
plt.xlabel("Ground truth length")
plt.ylabel("Count")
plt.show()
```
