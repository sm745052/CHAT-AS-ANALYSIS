```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
target = 0
eps = 1.5
```

```python
#compute predicted length distribution and gt length distribution
costs = []
predlen_dist = {}
gtlen_dist = {}
c=0

target_queries = []
with open('../eval.qb', 'r') as f:
    for query in f:
        query = query.strip()
        query = query.split('\t')
        if(len(query)<3):
            continue
        if(query[-1]=='-'):
            continue
        y = query[1]
        pred = query[2]
        predlen_dist[len(pred)] = predlen_dist.get(len(pred), 0) + 1
        gtlen_dist[len(y)] = gtlen_dist.get(len(y), 0) + 1
        if(float(query[4]) < 1e-6):
            c+= 1
            print(query)
            continue
        costs.append(float(query[3]) / float(query[4]))
        if(abs(costs[-1] - target) < eps and y != pred):
            target_queries.append(query)
```
